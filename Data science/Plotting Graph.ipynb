```python
import numpy as np
#x = np.array([1,2,3,4])
#color,marker,linewidth,labels for axis, dataset and title,axis,grid
x=np.arange(0,6,0.1)
y = x ** 3
y2 = x** 2
plt.plot(x,y,color = 'green', marker = "o", linewidth=1 , label = "x^3")
plt.plot(x,y2,color = 'red', marker = "*", linewidth=1,  label = "x^2")

#plt.axis([0,10,0,100])
plt.grid()
plt.xlabel("harsh")
plt.ylabel("agrawal")
plt.legend()
plt.title("Harsh AGrawal")
plt.text(2,80,"harsh")
```
